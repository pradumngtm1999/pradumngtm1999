Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Certainly! Seaborn is a powerful Python library for creating informative and attractive statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two numerical variables. They help in identifying patterns, clusters, or correlations between the variables.
Line Plot:

Use: Line plots are suitable for displaying trends over a continuous interval or time series data. They are used to show changes or trends in data over regular intervals.
Bar Plot:

Use: Bar plots are effective in comparing categorical data. They display the distribution or frequency of categorical variables and show the relationship between a numerical and a categorical variable.
Histogram:

Use: Histograms display the distribution of a single numerical variable. They help in visualizing the frequency or density of values within specific intervals (bins) to understand the data's underlying distribution.
Heatmap:

Use: Heatmaps are excellent for visualizing matrix-like data. They represent data values using a color scale and are commonly used in analyzing correlation matrices or representing relationships in large datasets.
Seaborn provides extensive support for these plots and offers customization options to enhance their visual representation and utility in data analysis and visualization.








Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset from Seaborn
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))  # Set the figure size

# Use Seaborn's lineplot function to create the plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)

# Set the title and labels
plt.title('FMRI Signal over Timepoints for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Show the plot
plt.legend(title='Event')  # Show legend with event categories
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset from Seaborn
titanic_data = sns.load_dataset("titanic")

# Create subplots for two box plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))  # Set the figure size

# Box plot for 'age' vs 'pclass'
sns.boxplot(x='pclass', y='age', data=titanic_data, ax=axes[0])
axes[0].set_title('Age distribution across Passenger Class')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')

# Box plot for 'fare' vs 'pclass'
sns.boxplot(x='pclass', y='fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Fare distribution across Passenger Class')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram with 'price' and differentiate by 'cut' using hue
plt.figure(figsize=(8, 6))  # Set the figure size

# Use Seaborn's histplot function to create the histogram
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)

# Set the title and labels
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.legend(title='Cut')  # Show legend with cut categories
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset from Seaborn
iris_data = sns.load_dataset("iris")

# Create a pair plot with 'species' as hue
plt.figure(figsize=(8, 6))  # Set the figure size

# Use Seaborn's pairplot function to create the pair plot
sns.pairplot(data=iris_data, hue='species')

# Set the title
plt.title('Pair Plot of Iris Dataset with Species Differentiation')

# Show the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset from Seaborn
flights_data = sns.load_dataset("flights")

# Rearrange the data into a pivot table to create a matrix for the heatmap
flights_pivot = flights_data.pivot('month', 'year', 'passengers')

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))  # Set the figure size

# Use Seaborn's heatmap function to create the heatmap
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')

# Set the title and labels
plt.title('Number of Passengers by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
