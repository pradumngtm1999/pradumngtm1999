Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.


Matplotlib is a comprehensive plotting library in Python used for creating static, interactive, and publication-quality visualizations. It provides a wide variety of plotting capabilities for visualizing data in different formats and styles.

Purpose of Matplotlib:
Data Visualization: Matplotlib enables users to create a diverse range of plots and graphs to visualize data, helping in data exploration, analysis, and presentation.
Publication-Quality Plots: It allows the creation of high-quality static visualizations suitable for publication in journals, reports, and presentations.
Customization: Matplotlib offers extensive customization options to control the appearance and style of plots, including colors, markers, labels, titles, and more.
Five plots that can be plotted using the Pyplot module of Matplotlib:
Line Plot: Represents data points connected by straight line segments.
Scatter Plot: Displays individual data points as markers to show the relationship between two variables.
Histogram: Shows the distribution of a dataset by dividing data into bins and displaying frequencies.
Bar Plot: Displays categorical data with rectangular bars of varying heights or lengths.
Pie Chart: Represents data as a circular statistical graphic divided into slices to illustrate numerical proportions.
Matplotlib's pyplot module provides a simple interface for creating various types of plots and customizing them to suit specific visualization requirements. It allows users to control almost every aspect of the plot, making it a powerful tool for data visualization in Python.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Creating the scatter plot
plt.scatter(x, y)

# Adding title, xlabel, and ylabel
plt.title('Scatter Plot of x and y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots and plotting four lines
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)  # Subplot 1
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)  # Subplot 2
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)  # Subplot 3
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)  # Subplot 4
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating a vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Vertical Bar Plot')
plt.show()

# Creating a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='salmon')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random data for box plots
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2])
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Values')
plt.title('Box Plot of Box1 and Box2')
plt.grid(True)
plt.show()
```
